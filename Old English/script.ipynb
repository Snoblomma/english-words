```python
f = open("Sources/Beowulf.txt", "r")

words = []

for line in f:
    entries = line.strip('\n\t"0123456789,.;:?!').split(" ")
    for entry in entries:
            entry = entry.strip('\n\t"0123456789,.;:?!').upper()
            if entry not in words:
                words.append(entry)
f.close()

words.sort()
len(words)
```
